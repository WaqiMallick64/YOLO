```python
# Import necessary libraries
import torch # PyTorch for deep learning computations
import cv2 # OpenCV for image processing
from ultralytics import YOLO # Ultralytics YOLOv8 model for object detection
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolov8s.pt")  # Load the YOLOv8s model can change it to depending on need
```

```python
# Upload an image from local system (specific to Google Colab)
from google.colab import files
uploaded = files.upload()  # Upload an image from your system

image_path = list(uploaded.keys())[0]
```

```python
results = model(image_path)  # Perform inference

# Display the image with bounding boxes
results[0].show()
```
